```python
import pandas as pd

df = pd.read_csv("../Planilhas/imoveis_unificado.csv", encoding='utf-8' ,sep=";")
df = df[df['logradouro'] != 'logradouro'].reset_index(drop=True)
df= df.drop_duplicates()
df = df.dropna()
```

```python

df["sfh"] = df["sfh"].astype(str).str.replace(",", ".", regex=False)
df["sfh"] = df["sfh"].astype(float)
df["sfh"].unique()
```

```python
df["area_terreno"] = df["area_terreno"].astype(str).str.replace(",", ".", regex=False)
df["area_terreno"] = df["area_terreno"].astype(float)
```

```python
df["tipo_construcao"] = df["tipo_construcao"].str.replace(" > 4 Pavimentos", "", regex=False).str.strip()
df["tipo_construcao"] = df["tipo_construcao"].str.replace(" <= 4 Pavimentos", "", regex=False).str.strip()
```
