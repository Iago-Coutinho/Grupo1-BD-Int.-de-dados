###Tratamento das colunas ano_construcao, cod_logradouro, valor_avaliacao###

Começando com a coluna ano_construcao, analisando as informações gerais de qual o estado inicial da coluna

```python
df1 = df['ano_construcao']
df1.info()

print(df['ano_construcao'].value_counts())

print(df['ano_construcao'].describe())
```

Como podemos ver por essas informações, teremos alguns pontos a serem tratados: O tipo de variável está incorreto, faz mais sentido estar como um int, por ser um ano; Já que não temos valores nulos, não precisamos nos preocupar com isso. O principal ponto de atenção são os anos sem coerência, como 670 (que são muitos antigos), ou valores que têm decimal. Felizmente, não tem valores acima de 2025, então nada sobresaiu o teto.

```python
print(df['ano_construcao'].unique())# Verifica o que está impedindo a conversão para int


df['ano_construcao'] = pd.to_numeric(df['ano_construcao'], errors='coerce').astype('Int64')# Converte forçando os erros para nulo 

print(df['ano_construcao'].dtype)
print(df['ano_construcao'].isna().sum())  # quantos viraram nulo

df['ano_construcao'].info()
```

Agora, irei tratar os valores nulos que não foram convertidos corretamente. Para repor esses valores nulos, vou usar a mediana do ano de construção das demais construções que estão no bairro daquela construção, para tornar mais coerente, mantendo as informações sobre bairros mais antigos ou mais novos. Também podemos ver que não tem nenhum valor absurdo como outlier, todas datas estão coerentes para um edifício atual

```python
df['ano_construcao'] = df['ano_construcao'].fillna(
    df.groupby('bairro')['ano_construcao'].transform('median') #comando usado para pegar a mediana dos anos de construção de um determinado bairro, e substituir nos locais nulos
)

df['ano_construcao'].info()
df['cod_logradouro'].head()
```
