# União dos arquivos CSV dos anos 2023, 2024, 2025 - ITBI Recife

Cria um único arquivo `imoveis_unificado.csv` com a união do conteúdo dos três anos.

> Algumas adaptações foram necessárias para que a união fosse executado de forma coerente.  

```python
# 2023 tem a coluna distrito
colunas_2023 = [
    'logradouro', 'numero', 'complemento', 'valor_avaliacao',
    'bairro', 'cidade', 'uf', 'ano_construcao',
    'area_terreno', 'area_construida', 'fracao_ideal',
    'padrao_acabamento', 'tipo_construcao', 'tipo_ocupacao',
    'data_transacao', 'estado_conservacao', 'tipo_imovel',
    'sfh', 'cod_logradouro', 'latitude', 'longitude', 'ano', 'distrito'
]

# 2024 e 2025 não têm distrito
colunas_2024_2025 = [
    'logradouro', 'numero', 'complemento', 'valor_avaliacao',
    'bairro', 'cidade', 'uf', 'ano_construcao',
    'area_terreno', 'area_construida', 'fracao_ideal',
    'padrao_acabamento', 'tipo_construcao', 'tipo_ocupacao',
    'data_transacao', 'estado_conservacao', 'tipo_imovel',
    'sfh', 'cod_logradouro', 'latitude', 'longitude', 'ano'
]



df_2023 = pd.read_csv('Planilhas/anos/2023.csv', sep=';', encoding='latin-1', low_memory=False)
df_2024 = pd.read_csv('Planilhas/anos/2024.csv', sep=';', encoding='latin-1', low_memory=False)
df_2025 = pd.read_csv('Planilhas/anos/2025.csv', sep=';', encoding='latin-1', low_memory=False)

# Renomear 'Distrito' para 'distrito' para padronizar
df_2023.rename(columns={'Distrito': 'distrito'}, inplace=True)
df_2024.rename(columns={'Distrito': 'distrito'}, inplace=True)

df = pd.concat([df_2023, df_2024, df_2025], ignore_index=True)
print(f"Total de linhas: {len(df)}")
df.head()
```

Transformação da união feita em um arquivo CSV

```python
df.to_csv('Planilhas/imoveis_unificado.csv', index=False, encoding='utf-8', sep=';')
print("Arquivo salvo com sucesso!")
```
